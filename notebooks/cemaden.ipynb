##Visualizando as estações no mapa

```python
def inline_map(m, width=650, height=500):
    """Takes a folium instance and embed HTML."""
    m._build_map()
    srcdoc = m.HTML.replace('"', '&quot;')
    embed = HTML('<iframe srcdoc="{}" '
                 'style="width: {}px; height: {}px; '
                 'border: none"></iframe>'.format(srcdoc, width, height))
    return embed
    

width, height = 650, 480
mapa_rj = folium.Map(location=[-22.9068, -43.1729], zoom_start=8, tiles='OpenStreetMap', width=width, height=height)
mapa_rj.lat_lng_popover()

ploted_stations = set([])
for row in dfrj.iterrows():
    lat = row[1]['latitude']
    lon = row[1]['longitude']
    if (lat,lon) in ploted_stations:
        continue
    nome = row[1]['nome']
    cod = row[1]['cod.estacao']
    marker = mapa_rj.circle_marker(location=[lat, lon], 
                                    fill_color='#f4bf75', 
                                    radius=22,
                                   popup='Nome: {}\nCodigo: {}'.format(nome, cod)
                                   )
    ploted_stations.add((lat,lon))
print len(ploted_stations)

inline_map(mapa_rj, width,height)
```
