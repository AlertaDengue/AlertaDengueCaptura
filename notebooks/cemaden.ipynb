```python
import requests
import pandas as pd
import codecs
```

```python
base_url="http://150.163.255.246:18383/dados_rede"
url_pcd = "http://150.163.255.246:18383/dados_pcd"
```

```python
chave = "bc10602ea62759fab1578f8eb1ff6f7abbf8678d"
```

#Testando a captura

```python
results = requests.get(base_url,params={'chave': chave,
                                        'inicio': 201412010000,
                                        'fim': 201412010300,
                                        'UF': "RJ"})
with codecs.open('clima.csv', 'w', encoding='utf8') as fd:
    fd.write(results.text)
```

Vamos agora puxar dados de uma estação específica:

```python
estacao = "120040101A"
results = requests.get(url_pcd,params={'chave': chave,
                                        'inicio': 201412010000,
                                        'fim': 201412012359,
                                        'codigo':"120040101A"})
with codecs.open('clima_{}.csv'.format(estacao), 'w', encoding='utf8') as fd:
    fd.write(results.text)
```
